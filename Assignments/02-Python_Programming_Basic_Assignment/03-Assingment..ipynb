```python
"""
1.	Write a Python Program to Check if a Number is Positive, Negative or Zero?

"""
```

```python
class numcheck:
    def __init__(self,num):
        self.num = num
    
    def check(self):
        if self.num > 0:
            return "Your number is positive"
        elif self.num < 0:
            return "your number is negetive"
        elif self.num == 0:
            return "your number is equal to zero"
        else:
            pass
```

```python
num1 = numcheck(10)
```

```python
num1.check()
```

```python
num2 = numcheck(-10)
```

```python
num2.check()
```

```python
num3 = numcheck(0)
```

```python
num3.check()
```

```python
"""
3.	Write a Python Program to Check Leap Year?

"""
```

```python
class leapyear:
    def __init__(self,year):
        self.year = year
    
    def leap(self):
        if (self.year % 400 == 0) and (self.year % 100 == 0):
            return "your enter year is leap year", self.year
        elif (self.year % 4 == 0) and (self.year % 100 != 0):
            return "your enter year is leap year", self.year
        else:
            return "your enter year isn't leap year", self.year
```

```python
l1 = leapyear(2020)
```

```python
l1.leap()
```

```python
l2 = leapyear(2022)
```

```python
l2.leap()
```

```python
"""
4.	Write a Python Program to Check Prime Number?

"""
```

```python
class Primenumber:
    def __init__(self,num):
        self.num = num
    
    def prime(self):
        if self.num > 1:
            for i in range(2,int(num)):
                if num % i == 0:
                    return "number is not prime", self.num
                    break
            else:
                return "number is prime", self.num
        else:
            "number is not prime", self.num
```

```python
p1 = Primenumber(23)
```

```python
p1.prime()
```

```python
p2 = Primenumber(25)
```

```python
p2.prime()
```
